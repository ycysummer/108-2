請問高嘉瑜發文最頻繁的時段是哪時候？

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Week'] = df['Date'].dt.strftime('%A')
df
```

```python
df_fish1 = df[df['Page_Name']=='高嘉瑜']

#算每星期每小時發了幾篇
df_fish1_post = df_fish1.groupby(['Week','Hour'])['LIKE_COUNT'].count().reset_index(name='count')

df_fish1_post
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

flights = df_fish1_post.pivot("Week", "Hour", "count")
# 畫出x-y-z的圖，比如 年-月-銷量 的熱圖
f, ax = plt.subplots(figsize=(9, 6))
#使用不同的顏色
sns.heatmap(flights, fmt="d",cmap='YlGnBu', ax=ax)
#設置座標字體方向
label_y = ax.get_yticklabels()
plt.setp(label_y, rotation=360, horizontalalignment='right')
label_x = ax.get_xticklabels()
plt.setp(label_x, rotation=45, horizontalalignment='right')
plt.show()
```

觀察：除了0時至6時之間的發文數比較少，其餘時間幾乎都至少有一篇貼文，可以說高嘉瑜是重度臉書使用者，利用臉書作為主要和選民分享訊息的工具。想比較看看與高嘉瑜性質相異的候選人林郁方（年紀較大、藍營、軍系、右派）的發文時段熱圖。

```python
df_lin1 = df[df['Page_Name']=='林郁方']

#算每星期每小時發了幾篇
df_lin1_post = df_lin1.groupby(['Week','Hour'])['LIKE_COUNT'].count().reset_index(name='count')

df_lin1_post
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

flights = df_lin1_post.pivot("Week", "Hour", "count")
# 畫出x-y-z的圖，比如 年-月-銷量 的熱圖
f, ax = plt.subplots(figsize=(9, 6))
#使用不同的顏色
sns.heatmap(flights, fmt="d",cmap='YlGnBu', ax=ax)
#設置座標字體方向
label_y = ax.get_yticklabels()
plt.setp(label_y, rotation=360, horizontalalignment='right')
label_x = ax.get_xticklabels()
plt.setp(label_x, rotation=45, horizontalalignment='right')
plt.show()
```

觀察：可以觀察到林郁方發文數不管是什麼時段都遠遠低於高嘉瑜，頻繁的發文時間落在21時至23時，推測這段時間是小編的上班時間，或者也有可能林郁方本人跑完行程之後的空檔自己發文。
