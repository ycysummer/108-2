```
import pandas as pd
df = pd.read_csv("C:/Users/Summer/Desktop/x/Taipei.csv")
df
```

讀取並顯示資料

```
name = df.groupby("Page_Name")
```

將name賦予df以page_name為主要排序的資料

```
p1 = name.get_group("高嘉瑜")['LIKE_COUNT'].agg('sum')
```

```
p2 = name.get_group("李彥秀")['LIKE_COUNT'].agg('sum')
```

```
data = {'Name':['KAO', 'LEE',],
        'LIKE':[p1, p2]}
```

將高嘉瑜及李彥秀的讚數總和置入data

```
likedata = pd.DataFrame(data)
```

將data以數據框方式存入likedata

```
likedata
```

顯示likedata

```
pc = pd.DataFrame({'LIKE': [p1, p2]},
                  index=['KAO', 'LEE'])
plot = pc.plot.pie(y='LIKE', figsize=(5, 5))
```
